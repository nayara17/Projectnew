# Loan Analysis Project Introduction
Welcome to the Loan Analysis project, where we dive into the world of peer-to-peer lending with a focus on risk analytics and machine learning. In this project, we will explore the data provided by LendingClub, the world's largest peer-to-peer lending platform, to understand how data-driven insights can help minimize the risks associated with lending to customers. Our goal is to develop a comprehensive understanding of the factors that influence loan default and leverage this knowledge to make informed lending decisions.

 When the company receives a loan application, the company has to make a decision for loan approval based on the applicantâ€™s profile. Two types of risks are associated with the bankâ€™s decision:

> - If the applicant is likely to repay the loan, then not approving the loan results in a loss of business to the company
> - If the applicant is not likely to repay the loan, i.e. he/she is likely to default, then approving the loan may lead to a financial loss for the company


> The data given contains the information about past loan applicants and whether they â€˜defaultedâ€™ or not. The aim is to identify patterns which indicate if a person is likely to default, which may be used for takin actions such as denying the loan, reducing the amount of loan, lending (to risky applicants) at a higher interest rate, etc.  

> When a person applies for a loan, there are two types of decisions that could be taken by the company:
> 1. `Loan accepted`: If the company approves the loan, there are 3 possible scenarios described below:
    - `Fully paid`: Applicant has fully paid the loan (the principal and the interest rate)
    - `Current`: Applicant is in the process of paying the instalments, i.e. the tenure of the loan is not yet completed. These candidates are not labelled as 'defaulted'.
    - `Charged-off`: Applicant has not paid the instalments in due time for a long period of time, i.e. he/she has defaulted on the loan
> 2. `Loan rejected`: The company had rejected the loan (because the candidate does not meet their requirements etc.). Since the loan was rejected, there is no transactional history of those applicants with the company and so this data is not available with the company (and thus in this dataset)


Project Summary

The core of the Loan Analysis project revolves around the evaluation of loan applications received by LendingClub. When a loan application is submitted, the company faces a crucial decision: whether to approve or reject the application. This decision is fraught with two types of risks:

Risk of Loss due to Non-Repayment: If the applicant is likely to repay the loan, denying the loan results in a loss of potential business for the company.

Risk of Default: If the applicant is unlikely to repay the loan, approving the loan may lead to a significant financial loss for the company if the applicant defaults.

To address these risks effectively, we need to analyze historical data on loan applicants and their repayment behaviors. The data contains information about applicants who were approved for loans and whether they eventually repaid the loan, are still in the repayment process, or defaulted.

Project Objectives


Our primary objectives in this project are as follows:

Risk Assessment: We aim to identify patterns and factors that indicate whether a loan applicant is likely to default. By doing so, we can take appropriate actions such as denying loans, reducing loan amounts, or offering loans to risky applicants at higher interest rates.

Credit Loss Reduction: Similar to other lending companies, LendingClub faces credit loss, which occurs when borrowers fail to repay their loans. By identifying and mitigating the risk associated with risky loan applicants (labeled as 'charged-off' or 'defaulters'), we can significantly reduce credit loss.

Driving Factors Analysis: We will explore the variables and factors that strongly correlate with loan default. Understanding these key drivers of default will allow us to make more informed decisions regarding loan approvals and risk assessment.

Portfolio and Risk Assessment: The insights gained from our analysis will help the company better manage its loan portfolio and assess the level of risk associated with different types of loans and applicants.

Machine Learning: In addition to exploratory data analysis (EDA), we will leverage machine learning techniques to build predictive models that can automate the process of identifying risky loan applicants.

In summary, this project is not only a journey into the world of finance and risk analytics but also a practical application of data science and machine learning to solve real-world business challenges. By the end of this project, we aim to provide actionable insights and models that can assist LendingClub in making more informed and profitable lending decisions.

```python
def pub_rec(number):
    if number == 0.0:
        return 0
    else:
        return 1
    
def mort_acc(number):
    if number == 0.0:
        return 0
    elif number >= 1.0:
        return 1
    else:
        return number
    
def pub_rec_bankruptcies(number):
    if number == 0.0:
        return 0
    elif number >= 1.0:
        return 1
    else:
        return number
```
